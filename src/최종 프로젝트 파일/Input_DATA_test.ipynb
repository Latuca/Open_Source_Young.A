# 노드 생성:출발지/도착지(정훈)

```python
import pandas as pd
import numpy as np
#원하는 개수만큼 각 노드(거점명)을 무작위 생성
def generatelocate(n):
    #모든 거점명
    locations = ["숭실대입구","상도역","보라매역","남성역","이수역"]
    #랜덤으로 정해진 거점명이 저장될 곳
    randomlocations = []
    #입력합 값만큼 n번 반복해서 랜덤으로 거점 생성후 저장
    for i in range(n):
        variance = np.random.randint(5)
        randomlocations.append(locations[variance])
    return randomlocations
Dept=generatelocate(1000)
Dest=generatelocate(1000)
```

```python
#과제:생성 버튼 누를 때마다 임의의 시간 생성해서 csv파일로 만들기
#해결 절차:날짜/시간을 범위로 서 지정한 뒤, random패키지의 uniform모듈을 통해 해당 범위 사이의 값을 임의로 택해서 
#새 변수에 저장 후 해당 변수 csv파일로 출력
#패키지 임포트
import random
import datetime
import pandas as pd
import csv
from pandas import Series,DataFrame
import os
os.getcwd()
os.chdir(r'/content/drive/MyDrive/Colab Notebooks/src/')

#시간 범위지정
start_time=datetime.datetime(2022,11,25,15,00,00)
end_time=datetime.datetime(2022,11,25,15,15,00) #날짜 범위 조정 변수-> 원하는 날짜를 고정하고 시간만 난수 생성하기 가능

num_time=1000
time_list=[]
for _ in range(num_time):
    rdm_time=str(random.uniform(start_time,end_time))
    result=rdm_time[0:16]
    time_list.append(result)
    random.seed(result)

print(time_list)
```

# 성별 및 택시 종류 데이터(준희)

```python
# 0과 1이 무한히 있는 주머니에서 300개를 뽑고 0를 남자, 1을 여자라고 하고 남자와 여자의 수를 csv파일로 저장하시오.
import numpy as np
import pandas as pd
import random
np.random.seed(0)
a = np.random.choice([0,1],1000 ,p=[0.7, 0.3])
b = pd.DataFrame(a)
#0과 1로 나누기
c = b[b[0]==0]
d = b[b[0]==1]
#남자와 여자의 수
men=list()
for i in range(len(c)):
    men.append(0)
women=list()
for i in range(len(d)):
    women.append(1)
sex=men+women
random.shuffle(sex)
print('남자의 수 :', len(c))
print('여자의 수 :', len(d))
print(sex)
#남자의 수와 여자의 수를 엑셀 형식으로 저장하고 csv파일로 저장

```

```python
#0,1,2이 무한히 있는 주머니에서 0,1,2가 뽑힐 확률이 각각 0.7, 0.2, 0.1일때 300개를 뽑고 0을 일반 택시, 1를 벤,2를 고급 택시라고 하고 일반 택시, 벤, 고급 택시의 수를 csv파일로 저장하시오.

import numpy as np
import pandas as pd
import random
np.random.seed(0)
a = np.random.choice([0,1,2], 1000, p=[0.7,0.2,0.1])
b = pd.DataFrame(a)
#0,1,2로 나누기
c = b[b[0]==0]
d = b[b[0]==1]
e = b[b[0]==2]
#일반 택시, 벤, 고급 택시의 수
normal=list()
for i in range(len(c)):
    normal.append(0)
ben=list()
for i in range(len(d)):
    ben.append(1)
superb=list()
for i in range(len(e)):
    superb.append(2)
taxi_type=normal+ben+superb
random.shuffle(taxi_type)
print('일반 택시의 수 :', len(c))
print('벤의 수 :', len(d))
print('고급 택시의 수 :', len(e))
print(taxi_type)
#일반 택시, 벤, 고급 택시의 수를 엑셀 형식으로 저장하고 csv파일로 저장
```
