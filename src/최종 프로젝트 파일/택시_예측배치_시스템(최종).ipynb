```python
import tensorflow as tf
import numpy as np
import pandas as pd
import random
```

```python
#시간 0~60분 생성
#0~30분은 가중치 0.7 30~60분은 가중치 0.3 부여
def time(n):
    timeline = []
    #생성후 각각 몇 개가 생성되었는지 확인하기 위해 만들었음
    global count_30분
    global count_60분
    #0~30분에 가중치 0.7,30~60분에 0.3부여
    for i in range(n):
        j = [0,1]
        k = np.random.choice(j, n, p=[0.7,0.3])
    #시간을 분단위로 쪼갬.
    for m in range(n):
        if k[m]==0:
            timeline.append(random.randint(0,31))
            count_30분 += 1
        else:
            timeline.append(random.randint(31,61))
            count_60분 += 1
    return timeline
        
```

```python
#택시종류 생성 
#일반=0, 고급=1
#각각 가중치 0.9,0.1 부여
def taxi(n):
  #생성후 각각 몇 개가 생성되었는지 확인하기 위해 만들었음
  global count_택시종류_일반
  global count_택시종류_고급
  택시 = [0,1]
  택시종류 = np.random.choice(택시,n,p=[0.9,0.1])
  #각 값이 얼마나 생성되었는지 카운트하였음
  for i in range(n):
    if 택시종류[i]==0:
      count_택시종류_일반 +=1
    else:
      count_택시종류_고급 +=1
  return 택시종류
```

```python
#출발지 생성
#노량진=1, 숭실대입구=100, 장승배기 = 200
#각각 가중치 0.7,0.2,0.1 부여
def depart(n):
  global count_노량진
  global count_숭실대입구
  global count_장승배기
  거점 = [1, 100, 200]
  출발지 = np.random.choice(거점,n,p=[0.7,0.2,0.1])
  #각 값이 얼마나 생성되었는지 카운트하였음
  for i in range(n):
    if 출발지[i] == 거점[0]:
      count_노량진 +=1
    elif 출발지[i] == 거점[1]:
      count_숭실대입구 +=1
    elif 출발지[i] == 거점[2]:
      count_장승배기 +=1
    else:
      count_남성 +=1
  return 출발지
```

```python
#데이터 n개 만들기
def generatedata(n):
  #시간, 택시종류 데이터 생성(x데이터에 들어갈 값)
  x데이터=[]
  시간생성 = time(n)
  택시생성 = taxi(n)
  for i in range(n):
    x데이터.append([시간생성[i],택시생성[i]])
  #y데이터에는 출발지가 들어감
  y데이터 = depart(n)
  return x데이터, y데이터
```

```python
count_30분 = 0
count_60분 = 0
count_택시종류_일반 = 0
count_택시종류_고급 = 0
count_노량진 = 0
count_숭실대입구 = 0
count_장승배기 =0
n = int(input("몇 개의 데이터를 생성 하시겠습니까?"))
x, y = generatedata(n)
print("0-30분",count_30분)
print("30-60분",count_60분)
print("일반",count_택시종류_일반)
print("고급",count_택시종류_고급)
print("노량진",count_노량진)
print("숭실대입구",count_숭실대입구)
print("장승배기",count_장승배기)
```

```python
#은닉층들 설정
model = tf.keras.models.Sequential([
    #은닉층 생성과 설정하기(노드(?)개수, 활성화 함수 입력)
    #마지막층은 항상 1로 해야하고 sigmoid라는 활성화 함수를 써야함
    tf.keras.layers.Dense(64, activation='tanh'),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(1, activation = 'sigmoid'),
])
```

```python
#모델 컴피일 설정. 최적화 모듈은 adam이라는 것을 이용했고 손실함수는 mse라는 회귀모델에서 많이 사용되는 함수, 그리고 마지막은 학습한 ai의 정확도를 출력해줌.
model.compile(optimizer='adam', loss='mse', metrics=['accuracy'])
```

```python
#데이터를 사용하기 전에 항상 데이터를 넘파이 어레이로 바꿔줘야함. epochs=''에 원하는 학습횟수 입력
#항상 많이 학습한다고 해서 좋은 건 아니니 적당히 조절해 가면서 학습 시켜주는 것이 좋음
model.fit( np.array(x), y, epochs=100)
```
